```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = np.arange(0,10)
plt.scatter(x,y, c='r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('scatter plot')
plt.savefig('scatterplot.png')
plt.show()

```

```python
plt.subplot(2,2,1)
plt.plot(x, y, 'r--')
plt.subplot(2,2,2)
plt.plot(x,y, 'g--')
plt.subplot(2,2,3)
plt.plot(x,y, 'bo')
plt.subplot(2,2,4)
plt.plot(x,y, 'go')
```

```python
x = np.arange(1,11)
y =( 3 * x) + 5
plt.scatter(x,y,c='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("Scatter plot")
plt.savefig('scatterplots.png')
```

```python
x= np.arange(0,4 * np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,2,1)
plt.plot(x,y_sin,'r')
plt.title("Sin plot")
plt.savefig('sin plots.png')
plt.subplot(2,2,2)
plt.plot(x,y_cos,'g')
plt.title("cos plot")
plt.savefig('cos.png')

```

```python
x = [2,8,10]
y = [11,16,9]
x2 = [3,9,11]
y2 = [6,15,12]
plt.bar(x,y)
plt.bar(x2, y2,color='r')
```

```python
x=np.array([22,11,47,52,12,60,8,27,42])
plt.hist(x)
plt.show()
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,)
plt.show()
```
